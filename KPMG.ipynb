```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#READ THE EXCEL FILE AND PRINT THE DATAHEAD OF TRANSACTION DATA
trn = pd.read_excel(r'C:\Users\User\Desktop\Data science internship\KPMG\kpmg_raw.xlsx',sheet_name = 1)
trn.head()
```

```python
#DESCRIBE THE DATA
trn.describe()
```

```python
#DROP THE DUPLICATES
trn = trn.drop_duplicates(subset = 'customer_id')
trn.describe()
```

```python
trn.head()
```

```python
trn['profit'] = trn['list_price'] - trn['standard_cost']
trn.head()
```

```python
trn = trn[['customer_id','profit']]
trn.head()
```

```python
trn = trn.sort_values(by ='profit' , ascending=False)
trn
```

```python
trn = trn[:750]
trn.describe()
```

```python
#READ THE EXCEL FILE AND PRINT THE DATAHEAD OF CUSTOMER DEMOGRAPHIC
dmg = pd.read_excel(r'C:\Users\User\Desktop\Data science internship\KPMG\kpmg_raw.xlsx',sheet_name = 3)
dmg.reset_index(drop=True, inplace=True)
dmg.head()
```

```python
dmg = dmg[['customer_id','gender','past_3_years_bike_related_purchases','Age','owns_car']]
dmg.head()
```

```python
dmg = dmg.replace(to_replace = 'F',value = 'Female')
dmg = dmg.replace(to_replace = 'M',value = 'Male')
dmg.head()
```

```python
#READ THE EXCEL FILE AND PRINT THE DATAHEAD OF CUSTOMER DEMOGRAPHIC
adr = pd.read_excel(r'C:\Users\User\Desktop\Data science internship\KPMG\kpmg_raw.xlsx',sheet_name = 4)
adr.reset_index(drop=True, inplace=True)
adr.head()
```

```python
adr = adr.replace(to_replace = 'Victoria',value = 'VIC')
adr = adr.replace(to_replace = 'New South Wales',value = 'NSW')
#adr.head()
```

```python
adr = adr[['customer_id','state']]
adr.head()
```

```python
#to print values in 'adr' dataframe according to 'customer_id' of 'trn' dataframe
options = trn['customer_id']
adr = adr[adr['customer_id'].isin(options)]
adr.head(100)
```

```python
adr.describe()
```

```python
#to print values in 'dmg' dataframe according to 'customer_id' of 'trn' & 'adr' dataframes
options = trn['customer_id']
dmg = dmg[dmg['customer_id'].isin(options)]
dmg.head(100)
```

```python
trn_adr = pd.merge(trn, adr, how='inner', on=None, left_on=None, right_on=None,
         left_index=False, right_index=False, sort=True,
         suffixes=('_x', '_y'), copy=True, indicator=False,
         validate=None)
trn_adr
```

```python
dmg
```

```python
#MERGING ALL DATA
final_data = pd.merge(dmg,trn_adr,how = 'right')
final_data
```

```python
final_data = final_data.sort_values(by ='profit' , ascending=False)
final_data
```

```python
# Printing the desired file in csv format
final_data.to_csv(r'C:\Users\User\Desktop\Data science internship\KPMG\kpmg_final_data.csv')
```
